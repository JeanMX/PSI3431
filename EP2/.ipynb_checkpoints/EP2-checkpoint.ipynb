```julia
using Plots, DSP, LinearAlgebra, LaTeXStrings
```

```julia
N = 101
n = 0:N - 1
L = (N - 1)  2
c = /4
h = (c/) * sinc.( (c / ) * (n .- L))

plot(n, h, line = :stem, marker = (:circle, 3), xlabel = "n", label = "h[n]", color = :magenta)

```

```julia
hdf = PolynomialRatio(h, [1])
 = range(0, , length = 2500)
Hd = freqresp(hdf, )
plot( / , abs.(Hd), label = "|Hd|", xlabel = "/")
```

```julia
n2 = 0:N - 1
h2 = ((-c)/) * sinc.( ((-c)/) * (n2 .- L)) .* (-1).^(n2)
plot(n2, h2, line = :stem, marker = (:circle, 3), xlabel = "n", label = "h2[n]", color = :magenta)
```

```julia
hdf2 = PolynomialRatio(h2, [1])
 = range(0, , length = 2500)
Hd2 = freqresp(hdf2, )
plot( / , abs.(Hd2), label = "|Hd2|", xlabel = "/")
```

```julia
function filtro(h, x)
    y = zeros(length(x))
    N = length(h)
    k = length(x)
    for n in 1:k
        for i in 0:n-1
            y[n] = y[n] + h[i+1]*x[n-i]
        end
    end
    
    return y
            
end
    
```

```julia
x = cos.(*n/20) + cos.(*n/3);
plot(n, x)
```

```julia
y11 = filtro(h, x);
y12 = filt(h, [1], x)

plot(n, y11)
plot!(n, y12)
```

```julia
y11 = filtro(h2, x);
y12 = filt(h2, [1], x)

plot(n, y11)
plot!(n, y12)
```
